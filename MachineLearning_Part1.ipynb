* bfill (Backward Fill): Fills NaN with the value from the next row (backward in time).

| Term       | Meaning / Direction                           | Example                                                                  |
| ---------- | --------------------------------------------- | ------------------------------------------------------------------------ |
| `df.shape` | Returns size of DataFrame → `(rows, columns)` | `df.shape → (3, 3)`                                                      |

| `axis=1`   | Operate **column-wise** (→ across rows)       | `df.dropna(axis=1)` → drop columns<br>`df.sum(axis=1)` → sum of each row |

Q2 is not used directly in IQR outlier detection, but it's useful for understanding the central tendency.

IQR focuses on Q1 and Q3, because outliers are defined as values outside:
* Q1 : 25th percentile (Lower Quartile).
* Q2 : 50th percentile (Median).
* Q3 : 75th percentile (Upper Quartile).

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
q1, q3
```

## Feature Scaling Techniques : 1. Standardization and 2. Normalization

| Type                | Method                        | Range            | When to Use                           |
| ------------------- | ----------------------------- | ---------------- | ------------------------------------- |
| **Normalization**   | Min-Max Scaling               | 0 to 1           | When data is not normally distributed |
| **Standardization** | Z-score scaling               | Mean = 0, SD = 1 | When data is normally distributed     |
| **Robust Scaling**  | Based on median and IQR       | Handles outliers | When data has outliers                |
| **MaxAbs Scaling**  | Divides by max absolute value | -1 to 1          | For sparse data or non-negative data  |

Feature Scaling is the process of adjusting the values of features (columns) in your dataset so they all lie in the same range. Feature scaling changes values into a similar range. It helps models work better by treating all features equally.

## Types of Ml :

- `Supervised learning` is a type of machine learning where the model is trained using labeled data. It learns to map inputs to known outputs and is commonly used for tasks like classification and regression. Example: Predicting house prices based on features like size, location, and number of rooms.

- `Unsupervised learning` deals with unlabeled data. The model tries to find hidden patterns or groupings within the data, such as in clustering or dimensionality reduction. Example: Grouping customers into segments based on purchasing behavior.

- `Semi-supervised learning` uses a small amount of labeled data along with a large amount of unlabeled data. This approach helps improve learning performance when labeled data is scarce. Example: Classifying medical images when only a few images are labeled.

- `Reinforcement learning` is based on learning through trial and error. An agent interacts with an environment and learns to make decisions by receiving rewards or penalties based on its actions. Example: A robot learning to walk by adjusting its movements based on success or failure.

`Linear models` assume a straight-line relationship between input variables (features) and the output.  
They work well when the data follows a linear trend. `Linear Regression, Multi-linear Regression, Lasso & Ridge Regression`.  
- Example: Linear Regression — predicting salary based on years of experience.

`Non-linear models` can capture complex relationships where data does not follow a straight line.  
They are used when the relationship between input and output is curved or more complex. `Polynomial Regression, Decision Tree Regression, Random Forest Regression, Support Vector Machine, K-Nearest Neighbour`. 
- Example: Decision Tree — classifying whether a customer will buy a product based on behavior and age.


## Types of Linear Regression

- Linear Regression is a supervised learning algorithm used to model the relationship between a dependent variable and one or more independent variables.

- `Simple Linear Regression` involves one independent variable and one dependent variable.  
Example: Predicting house price based on only its area.

- `Multiple Linear Regression` uses two or more independent variables to predict a dependent variable.  
Example: Predicting house price based on area, number of rooms, and location.

- Ridge Regression(L2) is a regularized version of linear regression that adds a penalty to the coefficients to prevent overfitting.  
It works well when there is multicollinearity among input features.

- Lasso Regression(L1) also adds a penalty, but it can shrink some coefficients to zero, effectively performing feature selection.  
It is useful when we want a simpler model with fewer features.

- Polynomial Regression is a form of linear regression where the relationship between variables is modeled as an nth-degree polynomial.  
It can fit non-linear data by adding polynomial terms to the model.

## Classification:
Classification is a supervised learning task where the model predicts the category (label/class) of given data based on past labeled data.

Types of Classification:
- Binary Classification - The model predicts **one of two possible classes**. Ex: Email classification (Spam or Not Spam).
- Multi-Class Classification - The model predicts **one class out of three or more categories**. Ex: Handwritten digit recognition (digits *0–9*).
- Multi-label Classification - A single instance can belong to **multiple classes at the same time**. Ex: A news article can be tagged as *“Politics”*, *“Economy”*, and *“World”*.

To evaluate Classification models we can make use of Log loss or Cross Entropy loss, Confusion Matrix, AUC-ROC Curve.

## Logistic Regression  
A linear model that comes under **Supervised Learning**. It is used for predicting a **categorical dependent variable** using a given set of **independent variables**. The output can be like Yes or No, 0 or 1,True or False.

1. **Binomial Logistic Regression**  Two possible outcomes. Ex: Email Classification Spam (1) or Not Spam (0).  

2. **Multinomial Logistic Regression**  More than two unordered categories. Ex: Fruit classification Apple, Mango, Banana.  

3. **Ordinal Logistic Regression**  More than two ordered categories. Ex: Movie ratings Poor, Average, Good, Excellent.

## Naive Bayes:
Naive Bayes is a classification algorithm that uses Bayes' Theorem to predict the class of data. It assumes that all features are independent, even though that may not be true (hence "naive"). Naive means lacking experience. Best for: 
spam filtering, text classification, etc. 

Types of Naive Bayes Classifiers: 
1. Gaussian Naive Bayes: Assumes that features follow a normal (Gaussian) distribution. Used when data is continuous (e.g., age, salary). 
2. Multinomial Naive Bayes: Used for discrete count data, like word frequencies. Common in text classification (e.g., spam detection). 
3. Bernoulli Naive Bayes: Works with binary/boolean features (e.g., word present or not). Also used in text data, especially for binary term presence.
